### 11 利用数组进行数据处理
#### 利用数组表达式替换循环的做饭，通常被称为矢量化
矢量化数组运算效率较高

```python
points=np.arange(-5,5,0.1)
points
```

```python
xs,ys=np.meshgrid(points,points)
xs
```

```python
import matplotlib.pyplot as plt
z=np.sqrt(xs**2+ys**2)
z
plt.imshow(z,cmap=plt.cm.gray);plt.colorbar()
plt.title("sss")
```

### 12 将条件逻辑表达为数组运算
#### numpy.where 函数是三元表达式 x if condition else y 的矢量化版本

```python
x=np.array([1.1,1.2,1.3,1.4,1.5])
y=np.array([2.1,2.2,2.3,2.4,2.5])
cond=np.array([True,False,True,True,False])
# 在满足cond的条件下时输出x，否则输出y
result=np.where(cond,x,y)
result
```

####np.where通常用于根据另一个数组而产生一个新的数组

#### 对于一个随机数组，将所有正值替换为2，将所有负值替换为-2

```python
arr=np.random.randn(4,4)
arr
```

```python
# 将所有的正值替换为2，将所有的负值替换为02
np.where(arr>0,2,-2)
```

```python
# 只将正值替换为2
np.where(arr>0,2,arr)
```

### 实例：随机漫步


```python
import random
position=0
walk=[position]
steps=1000
# 注意在py3中py2中的xrange由range所替代
# range（steps) 从0开始生成到steps的数组
xr=range(steps)
for i in xr:
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)
walk
```

```python
import numpy as np
# 实现随机漫步的方式二
nsteps=1000
draws=np.random.randint(0,2,size=nsteps)
steps=np.where(draws>0,1,-1)
# 求所有元素的累计和
# 数组当前值于之前值相加的和，组成一个新的数组
walk=steps.cumsum()
print(walk)
print(walk.min())
print(walk.max())
```

#### 一次模拟多个随机漫步

```python
nwalks=5000
nsteps=1000
draws=np.random.randint(0,2,size=(nwalks,nsteps))
steps=np.where(draws>0,1,-1)
walks=steps.cumsum(1)
print(walks)
```
