#### 1-2 dt的index
该月第一天至最后一天，时间间隔1小时

```python
import datetime
from dateutil.parser import parse
```

```python
start_datetime=datetime.datetime(2017,11,29,00)
start_datetime=datetime.date(start_datetime.year,start_datetime.month,1)  
start_datetime=parse(str(start_datetime))
start_datetime
```

找到当月最后一天

```python
end_datetime=pd.date_range(start_datetime,periods=1,freq='M')[0]
#end_datetime=end_datetime[0]+ datetime.timedelta(minutes=-1)
str(end_datetime)
end_datetime=datetime.datetime(end_datetime.year,end_datetime.month,end_datetime.day,23,59)
end_datetime=parse(str(end_datetime))
end_datetime
```

#### 生成当月的所有的时间列表，间隔为1小时

```python
date_list=pd.date_range(start_datetime,end_datetime,freq='H')
```

```python
date_list
```

#### 生成一个空的dataframe

```python
pd.DataFrame(columns=columns,index=date_list)
```
