```python
# 结果文件所在路径
filepath=r"201711.csv"
f = open(filepath)
columns=['AT','BP','DT','HU','RN','SL','VB','WD','WL','WS','WT']
result=pd.read_csv(f,index_col=0,parse_dates=True)
result.head()
```

```python
se=result['AT']
se.head()
```

```python
se['2017/11/1'].head()
```

```python
se['2017/11/1'].index
```

```python
se['2017/11/1'].values
```

```python
from pyecharts import Line
```

```python

attr = se['2017/11/1'].index
v1 = se['2017/11/1'].values
line = Line("气温")
line.add("气温", attr, v1, mark_point=["average"])
# line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average"])
line.show_config()
line.render()
```

```python
attr = se.index
v1 = se.values
line = Line("三亚11月气温")
line.add("气温", attr, v1, mark_point=["max","min"],mark_line=["average"])
# line.add("商家B", attr, v2, is_smooth=True, mark_line=["max", "average"])
line.show_config()
line.render()
```
