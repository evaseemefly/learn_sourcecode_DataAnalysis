```python
# load datasets
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

df_08 =pd.read_csv('clean_08.csv')
df_18 =pd.read_csv('clean_18.csv')
```

### 二、从 2008 年到 2018 年，综合 mpg 的分布是如何变化的？

```python
# 设置绘图的区域大小
fig_mpg, axes_mpg = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
# 使用sns.distplot绘制直方图
sns.distplot(df_08['cmb_mpg'], ax = axes_mpg[0])
_ = axes_mpg[0].set_title('08')
sns.distplot(df_18['cmb_mpg'], ax = axes_mpg[1])
_ = axes_mpg[1].set_title('18')
```

** 答案：** 变得更偏向右边

### 三、描述排量和综合 mpg 之间的相关性。

```python
fig_dspl_mpg, axes_dspl_mpg = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
sns.regplot(x=df_08['cmb_mpg'],y=df_08['displ'],ax=axes_dspl_mpg[0])
_ = axes_dspl_mpg[0].set_title('08')
sns.regplot(x=df_18['cmb_mpg'],y=df_18['displ'],ax=axes_dspl_mpg[1])
_ = axes_dspl_mpg[1].set_title('18')
```

### 四、描述温室气体得分与综合 mpg 之间的相关性。

```python
fig_green_mpg, axes_green_mpg = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
sns.regplot(x=df_08['cmb_mpg'],y=df_08['greenhouse_gas_score'],ax=axes_green_mpg[0])
_ = axes_green_mpg[0].set_title('08')
sns.regplot(x=df_18['cmb_mpg'],y=df_18['greenhouse_gas_score'],ax=axes_green_mpg[1])
_ = axes_green_mpg[1].set_title('18')
```
